### 单晶硅输出特性

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
V = np.array([0, 0.2, 0.4, 0.6, 0.8, 1, 1.2, 1.4, 1.6, 1.8, 2.0, 2.2, 2.32])  # 电压数据
I_out_1 = np.array([15.8, 15.8, 15.8, 15.7, 15.7, 15.7, 15.6, 15.4, 14.8, 13.4, 10.2, 5.7, 0.024])  # 单晶硅电流数据
```

```python
P_1 = V * I_out_1
max_power_1 = np.max(P_1)
max_power_index_1 = np.argmax(P_1)
optimal_voltage_1 = V[max_power_index_1]
optimal_current_1 = I_out_1[max_power_index_1]
optimal_resistance_1 = optimal_voltage_1 / optimal_current_1 *1000

plt.figure(figsize=(8, 5))
plt.plot(V, I_out_1, marker='o', linestyle='-', color='#1f77b4', label='输出电流 I (mA)')
plt.scatter(V, I_out_1, color='#1f77b4')
plt.plot(V, P_1, marker='o', linestyle='-', color='#ff7f0e', label='输出功率 P (mW)')
plt.scatter(V, P_1, color='#ff7f0e')
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title('单晶硅输出特性', fontsize=14)
plt.xlabel('输出电压 U (V)', fontsize=12)
plt.ylabel('输出电流 I (mA) / 输出功率 P (mW)', fontsize=12)
plt.legend()
plt.show()

print(f"单晶硅的最大功率: {max_power_1:.3f} mW")
print(f"最佳匹配负载电阻: {optimal_resistance_1:.1f} Ω")
```
