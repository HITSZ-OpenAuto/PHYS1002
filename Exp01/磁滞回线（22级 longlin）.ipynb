## table 1-1

```python
U_x_1 = [-587, -233, -160, -120.0, -80, -53.3, 0, 46.7, 66.7, 140, 573, 233, 160, 140, 46.7, 13.3, 0, -20, -60, -160, -587] # mV
U_y_1 = [-52.4, -40.4, -29.2, -21.2, -10, 0, 17.2, 27.6, 31.6, 41.2, 54, 41.2, 31.6, 27.6, 0, -10, -14, -21.2, -29.2, -40.4, -52.4] # mV

L_1 = 0.130 # m
S_1 = 1.24e-4 # m^2
N_1 = 150
N_2 = 150
```

```python
H_1 = [] # A/m
B_1 = [] # T
HB_1 = []

for i in range(0, len(U_x_1)):
    h = N_1/(L_1*R_1)*U_x_1[i]/1000
    H_1.append(round(h, 2))
    b = (R_2*C)/(N_2*S_1)*U_y_1[i]/1000
    B_1.append(round(b, 2))

print(H_1)
print(B_1)
```

```python
plt.plot(H_1, B_1, c='#F4E869')
plt.scatter(H_1, B_1, c='#F4E869')
plt.xlabel('H(A/m)')
plt.ylabel('B(T)')
plt.grid(True)
plt.title('table 1-1')
plt.legend()
plt.show()
H_1.clear()
B_1.clear()
```
